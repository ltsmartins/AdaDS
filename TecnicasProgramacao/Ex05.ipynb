```python
import pandas as pd
```

# Exercicio 1

```python
df = pd.read_csv('./alunos3.csv', sep=';', decimal=',')
df
```

```python
df['media_provas'] = df[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']].mean(axis=1)
df
```

# Exercicio 2

```python
df[(df['media_provas'] == df['media_provas'].min()) | (df['media_provas'] == df['media_provas'].max())]
```

# Exercicio 3

```python
cadastro_alunos = pd.read_excel('./cadastro_alunos.xlsx')
cadastro_alunos
```

```python
df = pd.merge(df, cadastro_alunos, how='left', on='RA')
df
```

# Exercicio 4

```python
mask_fem = df['Sexo'] == 'Feminino'
mask_masc = df['Sexo'] == 'Masculino'
print(f"Média das médias do grupo Feminino: {df[mask_fem]['media_provas'].mean()}")
print(f"Média das médias do grupo Masculino: {df[mask_masc]['media_provas'].mean()}")
```

# Exercicio 5

```python
mask_nota = df['media_provas'] >= 7
print(f"Média de idade entre alunos com nota >= 7: {df[mask_nota]['Idade'].mean()}")
```

```python
df_final = df.groupby('Cidade').agg({'media_provas':'mean'}).reset_index()
df_final.loc[df_final['media_provas'].idxmax()]
```
