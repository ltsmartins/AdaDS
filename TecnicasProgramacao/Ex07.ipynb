## 7 - Manipule os dados a seguir:
7.1 - Criar coluna "Total" somando o valor de todos os ativos por investidor<br>
7.2 - Criar coluna "Media" com a média das notas 1 e 2<br>
7.3 - Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)<br>
7.4 - Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)<br>
7.5 - Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar a função map)<br>
7.6 - Transformar os dados da coluna Media em inteiro<br>
7.7 - Criar coluna "NPS" onde o investidor é classificado como promotor se a média das notas é maior ou igual a 80, como neutro se a média estiver entre 50 e 80, e como detrator se for abaixo de 50<br>

```python
#7.1
df_final['Total_Ativos'] = df_final[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05', 'Ativo06', 'Ativo07']].sum(axis=1)
df_final.head()
```

```python
#7.2
df_final['Media'] = df_final[['Nota1', 'Nota2']].mean(axis=1)
df_final.head()
```

```python
#7.3
df_final['Domain' ] = df_final['Email'].str.split('@', expand=True)[1]
df_final.head()
```

```python
#7.4
df_final.rename(columns={'Salario':'Renda'}, inplace=True)
df_final['Renda'] = df_final['Renda'].map(lambda x: x.split('$')[1])
df_final['Renda'] = df_final['Renda'].astype(float)
df_final['Renda'].dtype
```

```python
#7.5
display(df_final['Sexo'].value_counts()) # <- isso prova que não temos outros valores a não ser 'FEM' e 'MASC', então não precisamos de nenhum tratamento e podemos aplicar o map direto.

df_final['Sexo'] = df_final['Sexo'].map(lambda x: 'Feminino' if x == 'Fem' else 'Masculino')

df_final.head()
```

```python
#7.6
df_final['Media'] = df_final['Media'].round().astype(int)
df_final.head()
```

```python
#7.7
values = [-float('inf'), 50, 80, float('inf')]
label = ('Detrator', 'Neutro', 'Promotor')

df_final['NPS'] = pd.cut(x=df_final['Media'], bins=values, labels=label)
df_final.head()
```
